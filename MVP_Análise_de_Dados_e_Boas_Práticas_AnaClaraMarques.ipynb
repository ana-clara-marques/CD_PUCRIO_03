# Descrição do Problema

O conjunto de dados utilizado neste trabalho é composto por registros meteorológicos de estações do Instituto Nacional de Meteorologia (**INMET**), localizadas no estado do **Rio de Janeiro**, abrangendo o período de **2000 a 2025**. As variáveis disponíveis incluem medidas de temperatura, umidade, vento, radiação e precipitação.

Além disso, foi utilizado um segundo conjunto de dados contendo **informações geográficas das estações**, como latitude, longitude e altitude, permitindo incorporar características físicas relevantes à análise.

O objetivo principal deste trabalho é **investigar a relação entre altitude e variáveis de temperatura**, com foco na temperatura média e na **amplitude térmica diária**, nas diferentes estações meteorológicas do estado do Rio de Janeiro.

Busca-se analisar se existem diferenças significativas nos padrões de temperatura entre estações localizadas em diferentes faixas de altitude, bem como avaliar se a variação térmica ao longo do dia pode estar associada não apenas à altitude, mas também a fatores como a **umidade do ar**.

Adicionalmente, o trabalho enfatiza a importância das etapas de **verificação e preparação dos dados**, incluindo tratamento de inconsistências, conversão de tipos e validação de valores, como base para a realização de análises confiáveis.

A partir dessas etapas, são conduzidas análises exploratórias com o objetivo de identificar padrões, relações e comportamentos relevantes nos dados meteorológicos.


## Hipóteses do Problema

As hipóteses que orientam esta análise são as seguintes:

1. Existe uma **relação inversa entre altitude e temperatura média do ar**, de modo que estações localizadas em maiores altitudes tendem a apresentar temperaturas mais baixas.
2. Estações em **diferentes faixas de altitude** apresentam **distribuições distintas de temperatura média, máxima e mínima**, evidenciando padrões térmicos característicos.
3. Estações localizadas em **maiores altitudes tendem a apresentar maior amplitude térmica diária** (diferença entre temperatura máxima e mínima) em comparação às estações de menor altitude. Uma possível explicação para esse comportamento é a maior influência marítima nas regiões de baixa altitude no estado do Rio de Janeiro, associada a níveis mais elevados de umidade do ar, o que pode reduzir a variação térmica ao longo do dia.

## Limpeza e Preparação dos Dados

Com base na verificação inicial, foram identificadas inconsistências nos tipos de dados e possíveis valores fora de faixas plausíveis para variáveis meteorológicas.

Inicialmente, foram realizados ajustes estruturais, incluindo a padronização do nome da coluna de data e sua conversão para o tipo datetime, permitindo análises temporais. Também foi realizada a conversão das variáveis geográficas (latitude, longitude e altitude), originalmente armazenadas como texto, para formato numérico.

Em seguida, os datasets foram integrados por meio do identificador da estação, possibilitando incorporar a informação de altitude à base meteorológica.

Para garantir a qualidade dos dados, foi aplicado um tratamento de valores inconsistentes nas variáveis de interesse. Com base em registros históricos do estado do Rio de Janeiro, sabe-se que os valores de temperatura variam aproximadamente entre -3°C e 44°C. No caso da temperatura, considerou-se um intervalo plausível entre -10°C e 50°C, com base em registros históricos do estado do Rio de Janeiro. Valores fora desse intervalo foram considerados inconsistentes e substituídos por valores nulos.

Para a umidade relativa do ar, foram considerados válidos apenas valores entre 0% e 100%, sendo valores fora desse intervalo também tratados como inconsistentes.

Por fim, foi criada a variável de amplitude térmica diária, definida como a diferença entre a temperatura máxima e a temperatura mínima, e foram selecionadas as variáveis relevantes para a análise.

```python
# ===============================
# 1. Ajuste de nomes e data
# ===============================
df_weather = df_weather.rename(columns={
    "DATA (YYYY-MM-DD)": "DATA"
})

df_weather["DATA"] = pd.to_datetime(df_weather["DATA"])
df_weather["ANO"] = df_weather["DATA"].dt.year

# ===============================
# 2. Limpeza do dataset de estações
# ===============================
df_stations["lat"] = df_stations["lat"].str.replace(",", ".")
df_stations["lon"] = df_stations["lon"].str.replace(",", ".")
df_stations["lvl"] = df_stations["lvl"].str.replace(",", ".")

df_stations["lat"] = pd.to_numeric(df_stations["lat"])
df_stations["lon"] = pd.to_numeric(df_stations["lon"])
df_stations["lvl"] = pd.to_numeric(df_stations["lvl"])

# ===============================
# 3. Merge dos datasets
# ===============================
df = df_weather.merge(
    df_stations,
    left_on="ESTACAO",
    right_on="id_station",
    how="left"
)

# ===============================
# 4. Tratamento de valores inconsistentes
# ===============================

# Temperatura (-10 a 50 °C)
temp_cols = ["temp_avg", "temp_max", "temp_min"]
for col in temp_cols:
    df[col] = df[col].where((df[col] >= -10) & (df[col] <= 50), np.nan)

# Umidade (0 a 100 %)
hum_cols = ["hum_max", "hum_min"]
for col in hum_cols:
    df[col] = df[col].where((df[col] >= 0) & (df[col] <= 100), np.nan)

# ===============================
# 5. Criação de variável derivada
# ===============================
df["temp_range"] = df["temp_max"] - df["temp_min"]

# ===============================
# 6. Seleção de variáveis relevantes
# ===============================
df = df[[
    "ESTACAO",
    "DATA",
    "ANO",
    "lvl",
    "temp_avg",
    "temp_max",
    "temp_min",
    "temp_range",
    "hum_max",
    "hum_min"
]]

# ===============================
# 7. Verificação final após limpeza
# ===============================
print("Valores nulos após limpeza:")
print(df.isnull().sum())
```
